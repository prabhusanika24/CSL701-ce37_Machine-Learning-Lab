Phase 4: Data Manipulation & Cleaning
Task 9: Fill missing Age values with median

```python
df["Age"] = df["Age"].fillna(df["Age"].median())

print(df["Age"].isnull().sum())
```

Task 10: Filter surviving passengers

```python
survivors = df[df["Survived"] == 1]

survivors.head()
```

Task 11: Convert Fare column into NumPy array

```python
fare_array = df["Fare"].to_numpy()

print(fare_array)
```

Task 12: Maximum and Average Fare

```python
print("Maximum Fare:", np.max(fare_array))
print("Average Fare:", np.mean(fare_array))
```

Phase 5: Data Visualization
Task 13: Histogram of Age

```python
plt.figure(figsize=(6,4))
plt.hist(df["Age"], bins=20)
plt.title("Histogram of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

Task 14: Bar Chart of Survival Count

```python
survival_count = df["Survived"].value_counts()

plt.figure(figsize=(5,4))
plt.bar(survival_count.index.astype(str), survival_count.values)
plt.title("Survival Count")
plt.xlabel("Survived")
plt.ylabel("Count")
plt.show()
```

Task 15: Scatter Plot of Age vs Fare

```python
plt.figure(figsize=(6,4))
plt.scatter(df["Age"], df["Fare"])
plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
